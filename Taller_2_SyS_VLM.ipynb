- #### **Serie de Fourier**
- Se aplica a señales **periódicas** en **tiempo continuo**.
- Descompone una señal en **sumas de senos y cosenos** o **exponenciales complejas**.
- Su resultado es un **espectro discreto** de coeficientes (suma de armónicos a frecuencias específicas).
- Existen tres formas:
  - **Trigonométrica**: usa senos y cosenos.
  - **Exponencial**: usa complejos \( e^{jn\omega_0 t} \).
  - **Compacta**: forma resumida de las anteriores.
---


  #### **Transformada de Fourier (FT)**
- Se usa en señales **no periódicas** y **continuas**.
- Devuelve una función continua que representa **cómo varía la señal en frecuencia**.
- El espectro resultante también es **continuo**.
- La Transformada de Fourier puede verse como una generalización de la Serie de Fourier para señales no periódicas.

---

#### **Transformada de Fourier en Tiempo Discreto (DTFT)**
- Se aplica a señales **no periodicas, discretas e infinitas** (como una secuencia muestreada que no termina).
- Genera un **espectro continuo y periódico**, con período $ 2\pi $
- Se parece a la FT, pero adaptada al caso discreto, es decir, es el equivalente de la Transformada de Fourier para señales discretas en el tiempo.

---

#### **Transformada Discreta de Fourier (DFT)**
- Se aplica a señales **discretas y finitas**.
- Produce un **espectro discreto y finito**, ideal para trabajar con computadoras.
- Es una versión muestreada de la DTFT, la DFT muestrea el espectro continuo de la DTFT a frecuencias discretas. Es la herramienta que se utiliza en la práctica para analizar señales digitales debido a que opera con un número finito de muestras tanto en el tiempo como en la frecuencia.


---

### FFT (Fast Fourier Transform)

- La **FFT** es un algoritmo eficiente para calcular la DFT y su inversa.
- Mientras que la DFT normal hace muchos cálculos, la FFT usa menos pasos, lo que la hace mucho más rápida.
- Funciona dividiendo el problema en partes más pequeñas y resolviéndolas paso a paso.
- Es muy útil cuando se tienen muchos datos (como en señales de audio o imágenes), ya que ahorra mucho tiempo y permite procesar señales en tiempo real.
- Mientras que la DFT necesita **O(N²)** operaciones para una señal de N muestras (porque hace una multiplicación por cada par de puntos), la FFT reduce ese número a solo **O(N log N)**. Esta mejora es muy significativa cuando N es grande.

#### Utilidad de la FFT

- Permite **analizar señales rápidamente**, incluso en tiempo real.
- Se usa en procesamiento de audio, imágenes, comunicaciones, análisis de vibraciones, compresión, entre otros.
- Es la base de herramientas como el espectrograma, filtros digitales y ecualizadores.

#### Comparación del costo computacional:

| Método | Complejidad | Tiempo relativo |
|--------|-------------|-----------------|
| DFT directa | $(N^2)$ | Lento para señales grandes |
| FFT        | $N \log N$ | Mucho más rápido y eficiente |

**3. a)** $x(t) = e^{-a|t|},\quad a \in \mathbb{R}^{+}$

Como esta señal es par, podemos escribir:

$$
x(t) =
\begin{cases}
e^{-a t}, & t \geq 0 \\
e^{a t}, & t < 0
\end{cases}
$$

$$
X(\omega) = \int_{0}^{\infty} e^{-a t} e^{-j\omega t} dt + \int_{0}^{\infty} e^{-a t} e^{j\omega t} dt
$$

$$
= \int_{0}^{\infty} e^{-(a + j\omega)t} dt + \int_{0}^{\infty} e^{-(a - j\omega)t} dt
$$
$$
\int_{-\infty}^{0} \left( e^{(a - j\omega)t} \right) dt
= \int_{-\infty}^{0} e^{(a - j\omega)t} dt
= \left[ \frac{e^{(a - j\omega)t}}{a - j\omega} \right]_{-\infty}^{0}
= \frac{1}{a - j\omega}
$$

$$
\int_{0}^{\infty} e^{-(a + j\omega)t} dt
= \left[ \frac{e^{-(a + j\omega)t}}{-(a + j\omega)} \right]_{0}^{\infty}
= 0 - \left[ \frac{1}{-(a + j\omega)} \right]
= \frac{1}{a + j\omega}
$$

Entonces:

$$
X(\omega) = \frac{1}{a - j\omega} + \frac{1}{a + j\omega}
= \frac{2a}{a^2 + \omega^2}
$$

**3. b)** $ \cos(\omega_c t), \quad \omega_c \in \mathbb{R} $

$$
X(\omega) = \int_{-\infty}^{\infty} \cos(\omega_c t) e^{-j\omega t} \, dt
$$

$$
\cos(\omega_c t) = \frac{e^{j\omega_c t} + e^{-j\omega_c t}}{2}, \quad
X(\omega) = \int_{-\infty}^{\infty} \left( \frac{e^{j\omega_c t} + e^{-j\omega_c t}}{2} \right) e^{-j\omega t} \, dt
$$

$$
X(\omega) = \frac{1}{2} \int_{-\infty}^{\infty} e^{j(\omega_c - \omega)t} \, dt + \frac{1}{2} \int_{-\infty}^{\infty} e^{-j(\omega_c + \omega)t} \, dt
$$

Como la integral de una exponencial compleja es un delta de Dirac:

$$
\int_{-\infty}^{\infty} e^{j\alpha t} \, dt = 2\pi \delta(\alpha)
$$

Aplicando esto a cada término se tiene:

$$
X(\omega) = \frac{1}{2} \left( 2\pi \delta(\omega - \omega_c) + 2\pi \delta(\omega + \omega_c) \right)
$$

$$
X(\omega) = \pi \left[ \delta(\omega - \omega_c) + \delta(\omega + \omega_c) \right]
$$

**3. c)** $x(t) = \sin(\omega_s t), \quad \omega_s \in \mathbb{R}$

Aplicando la definición de la transformada de Fourier se tiene:

$$
X(\omega) = \int_{-\infty}^{\infty} \sin(\omega_s t) \cdot e^{-j\omega t} \, dt
$$

Aplicando la identidad trigonométrica:  
$\sin(\theta) = \frac{e^{j\theta} - e^{-j\theta}}{2j}$

$$
X(\omega) = \int_{-\infty}^{\infty} \frac{e^{j\omega_s t} - e^{-j\omega_s t}}{2j} \cdot e^{-j\omega t} \, dt
= \frac{1}{2j} \int_{-\infty}^{\infty} \left( e^{j\omega_s t} - e^{-j\omega_s t} \right) \cdot e^{-j\omega t} \, dt
$$

$$
X(\omega) = \frac{1}{2j} \left( \int_{-\infty}^{\infty} e^{-j(\omega - \omega_s)t} \, dt - \int_{-\infty}^{\infty} e^{-j(\omega + \omega_s)t} \, dt \right)
$$

Como la integral de una exponencial compleja es una delta de Dirac:

$$
\int_{-\infty}^{\infty} e^{j\alpha t} \, dt = 2\pi \delta(\alpha)
$$

- En la primera integral: $\alpha = \omega_s - \omega \Rightarrow 2\pi \delta(\omega - \omega_s)$  
- En la segunda integral: $\alpha = -(\omega + \omega_s) \Rightarrow 2\pi \delta(\omega + \omega_s)$

Entonces:

$$
X(\omega) = \frac{1}{2j} \left( 2\pi \delta(\omega - \omega_s) - 2\pi \delta(\omega + \omega_s) \right)
$$
$$
X(\omega) = \pi \left[ \delta(\omega - \omega_s) - \delta(\omega + \omega_s) \right] \cdot \frac{1}{j}
$$

$$
X(\omega) = \sqrt{\pi} \left[ \delta(\omega + \omega_s) - \delta(\omega - \omega_s) \right]
$$

**3. d)** $x(t) = f(t) \cdot \cos(\omega t), \quad \omega \in \mathbb{R}, \quad f(t) \in \mathbb{R}$

Aplicando la definición de la transformada de Fourier:

$$
X(\omega) = \int_{-\infty}^{\infty} f(t) \cdot \cos(\omega t) \cdot e^{-j\omega t} \, dt
$$

Y aplicando la identidad trigonométrica:

$$
\cos(\omega t) = \frac{1}{2} \left( e^{j\omega t} + e^{-j\omega t} \right)
$$

$$
X(\omega) = \int_{-\infty}^{\infty} f(t) \cdot \frac{1}{2} \left( e^{j\omega t} + e^{-j\omega t} \right) \cdot e^{-j\omega t} \, dt
= \frac{1}{2} \int_{-\infty}^{\infty} f(t) \cdot e^{-j(\omega - \omega_c)t} \, dt
+ \frac{1}{2} \int_{-\infty}^{\infty} f(t) \cdot e^{-j(\omega + \omega_c)t} \, dt
$$

$$
= \frac{1}{2} \left[ \mathscr{F}\{f(t)\}(\omega - \omega_c) + \mathscr{F}\{f(t)\}(\omega + \omega_c) \right]
$$

Se observa que cada una de esas integrales corresponde a la transformada de Fourier de $f(t)$ evaluada en frecuencias desplazadas.

Así que por definición:

$$
X(\omega) = \frac{1}{2} \left[ F(\omega - \omega_c) + F(\omega + \omega_c) \right], \quad F(\omega) = \mathscr{F}\{f(t)\}
$$

**3. e)** $x(t) = e^{-a|t|^2}, \quad a \in \mathbb{R}^+, \quad$ señal Gaussiana.

$$
X(\omega) = \int_{-\infty}^{\infty} e^{-a|t|^2} \cdot e^{-j\omega t} \, dt
$$

Pero como $|t|^2 = t^2$ para todo $t \in \mathbb{R}$

$$
X(\omega) = \int_{-\infty}^{\infty} e^{-a t^2} \cdot e^{-j\omega t} \, dt
= \int_{-\infty}^{\infty} e^{-a t^2 - j\omega t} \, dt
$$

Llevando $-a t^2 - j \omega t$ a una forma factorizada se tiene (término cuadrático completo): $(t^2 - bt)$
$$
-a t^2 - j\omega t = -a \left( t^2 + \frac{j\omega}{a} t \right)
\quad \text{completando el cuadrado de } t^2 + \frac{j\omega}{a}t
$$

$$
= -a \left[ \left( t + \frac{j\omega}{2a} \right)^2 - \left( \frac{j\omega}{2a} \right)^2 \right]
$$

Entonces:

$$
-a t^2 - j\omega t = -a \left( \left( t + \frac{j\omega}{2a} \right)^2 - \left( \frac{j\omega}{2a} \right)^2 \right)
$$

Recordando que $j^2 = -1$, queda:

$$
= -a \left( \left( t + \frac{j\omega}{2a} \right)^2 + \frac{\omega^2}{4a^2} \right)
= -a \left( t + \frac{j\omega}{2a} \right)^2 - \frac{\omega^2}{4a}
$$
Sustituyendo y simplificando en la integral:

$$
X(\omega) = \int_{-\infty}^{\infty} e^{-a\left(t + \frac{j\omega}{2a} \right)^2} \cdot e^{\frac{-\omega^2}{4a}} \, dt
$$

Separando la parte constante del exponente que no depende de $t$:

$$
X(\omega) = e^{\frac{-\omega^2}{4a}} \int_{-\infty}^{\infty} e^{-a\left(t + \frac{j\omega}{2a} \right)^2} \, dt
$$

Realizando un cambio de variable:

$$
u = t + \frac{j\omega}{2a} \quad \Rightarrow \quad du = dt
$$

Cuando $t \to -\infty \Rightarrow u \to -\infty$,  
cuando $t \to \infty \Rightarrow u \to \infty$

Entonces:

$$
X(\omega) = e^{\frac{-\omega^2}{4a}} \int_{-\infty}^{\infty} e^{-a u^2} \, du
$$

La integral gaussiana es una integral estándar:

$$
\int_{-\infty}^{\infty} e^{-c u^2} \, dx = \sqrt{\frac{\pi}{c}}
$$

En este caso, $c = a$ y la variable es $u$,  
por lo tanto:

$$
\int_{-\infty}^{\infty} e^{-a u^2} \, du = \sqrt{\frac{\pi}{a}} \quad \Rightarrow \quad X(\omega) = e^{\frac{-\omega^2}{4a}} \cdot \sqrt{\frac{\pi}{a}}
$$

$$
\cos(\omega_c t) = \frac{1}{2} \left( e^{j\omega_c t} + e^{-j\omega_c t} \right)
$$

Entonces:

$$
e^{-j\omega_1 t} \cos(\omega_c t) = \frac{1}{2} \left( e^{-j(\omega_1 - \omega_c)t} + e^{-j(\omega_1 + \omega_c)t} \right)
$$

Aplicando la transformada de Fourier:

$$
\mathscr{F}\{ e^{-j\omega_1 t} \cos(\omega_c t) \}

\mathscr{F}\{ e^{j\omega_0 t} \} = 2\pi \delta(\omega - \omega_0)

Por lo tanto:

$$
\mathscr{F}\{ e^{-j\omega_1 t} \cos(\omega_c t) \}
= \frac{1}{2} \left[ 2\pi \delta(\omega - (\omega_1 - \omega_c)) + 2\pi \delta(\omega - (\omega_1 + \omega_c)) \right]
$$

Resultado final:

$$
\mathscr{F}\{ e^{-j\omega_1 t} \cos(\omega_c t) \}
= \pi \left[ \delta(\omega - (\omega_1 - \omega_c)) + \delta(\omega - (\omega_1 + \omega_c)) \right]

\cos^2(\omega_c t) = \frac{1}{2} + \frac{1}{2} \cos(2\omega_c t)

Entonces:

$$
u(t) \cos^2(\omega_c t) = u(t) \left( \frac{1}{2} + \frac{1}{2} \cos(2\omega_c t) \right)
= \frac{1}{2} u(t) + \frac{1}{2} u(t) \cos(2\omega_c t)

$$
\mathscr{F} \{ u(t) \cos^2(\omega_c t) \}

\mathscr{F} \{ u(t) \} = \pi \delta(\omega) + \frac{1}{j\omega}

\mathscr{F} \{ u(t) \cos(\omega_0 t) \} = \frac{1}{2} \left[

$$
\mathscr{F} \{ u(t) \cos(2\omega_c t) \}

Resultado final:
$$
\mathscr{F} \left\{ u(t) \cos^2(\omega_c t) \right\}
= \frac{\pi}{2} \delta(\omega) + \frac{1}{2j\omega} + \frac{\pi}{4} \left[ \delta(\omega - 2\omega_c) + \delta(\omega + 2\omega_c) \right]
+ \frac{1}{4j} \left( \frac{1}{\omega - 2\omega_c} + \frac{1}{\omega + 2\omega_c} \right)
$$

**c).**  
$\mathscr{F} \left\{ \frac{7}{\omega^2 + 6\omega + 45} * \frac{10}{(8 + j\omega / 3)^2} \right\}$

Aplicando el teorema de convolución para la transformada de Fourier:

$$
\mathscr{F} \left\{ f(t) * g(t) \right\} = 2\pi \cdot F(\omega) \cdot G(\omega)
$$

Donde:

- $F(\omega) = \mathscr{F} \left\{ \frac{7}{\omega^2 + 6\omega + 45} \right\}$
- $G(\omega) = \mathscr{F} \left\{ \frac{10}{(8 + j\omega / 3)^2} \right\}$



**Calcular la transformada inversa de la primera función $F(\omega)$:**

$$
F(\omega) = \frac{7}{\omega^2 + 6\omega + 45}
$$

Factorizando el denominador:

$$
\omega^2 + 6\omega + 45 = (\omega + 3)^2 + 36
$$

Aplicando la propiedad par de transformada (desplazamiento exponencial):

$$
F(\omega) = \frac{7}{(\omega + 3)^2 + 6^2}
$$

Aplicamos la forma conocida:

$$
\mathscr{F} \left\{ e^{-a t} \cdot \sin(b t) \cdot u(t) \right\} = \frac{2b}{(a + j\omega)^2 + b^2}, \quad a = 0
$$

Entonces:

$$
\mathscr{F}^{-1} \left\{ \frac{2 \cdot 6}{\omega^2 + 6^2} \right\} = e^{-0t} \cdot \sin(6t) \cdot u(t) = \sin(6t) \cdot u(t)
$$

Ajustando las constantes:

$$
H(\omega) = \frac{12}{\omega^2 + 6^2} \quad \Rightarrow \quad h(t) = \mathscr{F}^{-1} \{ H(\omega) \} = \frac{12}{6} \sin(6t) u(t) = 2 \sin(6t) u(t)
$$

Luego, aplicando la propiedad de desplazamiento en frecuencia:

$$
H(\omega + 3) \quad \Rightarrow \quad \omega_3 = 0 \quad \Rightarrow \quad \omega_0 = -3
$$

$$
f(t) = \mathscr{F}^{-1} \{ H(\omega + 3) \} = e^{-j3t} \cdot h(t)
$$

Resultado:

$$
f(t) = \frac{7}{12} \cdot e^{j3t} \cdot e^{-j3t} \cdot \sin(6t) u(t)
$$


Calcular la transformada inversa de la segunda función $g(t)$:

$$
G(\omega) = \frac{10}{(8 + j\omega / 3)^2}
$$

Aplicando el par de transformada:

$$
\mathscr{F} \left\{ t e^{-a t} u(t) \right\} = \frac{1}{(a + j\omega)^2}
$$

Reescribiendo $G(\omega)$:

$$
G(\omega) = \frac{10}{\left( \frac{1}{3} (24 + j\omega) \right)^2} = \frac{10}{\frac{1}{9}(24 + j\omega)^2}
= \frac{90}{(24 + j\omega)^2}
$$
La expresión:

$$
\frac{90}{(24 + j\omega)^2}
$$

coincide con la forma:

$$
\frac{1}{(a + j\omega)^2}
$$

Entonces:

$$
g(t) = \mathscr{F}^{-1} \left\{ \frac{90}{(24 + j\omega)^2} \right\}
= 90 \cdot \mathscr{F}^{-1} \left\{ \frac{1}{(24 + j\omega)^2} \right\}
$$

Resultado:

$$
g(t) = 90 \cdot t \cdot e^{-24t} \cdot u(t)
$$

Aplicando el teorema de convolución:

$$
y(t) = 2\pi \cdot f(t) \cdot g(t)
= 2\pi \cdot \left( \frac{7}{12} e^{-j3t} \cdot e^{-6|t|} \right) \cdot \left( 90 t e^{-24t} u(t) \right)
$$

$$
y(t) = 105\pi \cdot e^{-j3t} \cdot e^{-6|t|} \cdot t \cdot e^{-24t} \cdot u(t)
$$

Recordando que:

- $u(t) = 0$ para $t < 0$
- $u(t) = 1$ para $t \geq 0$
- $|t| = t$ para $t \geq 0$

Entonces para $t \geq 0$:

$$
e^{-6|t|} \cdot e^{-24t} = e^{-6t} \cdot e^{-24t} = e^{-30t}
$$

Resultado final:

$$
y(t) = 105\pi \cdot t \cdot e^{-(30 + j3)t} \cdot u(t)
$$

**d).** $\mathscr{F} \{ 3 t^3 \}$

Aplicando la propiedad de linealidad:

$$
\mathscr{F} \{ 3 t^3 \} = 3 \cdot \mathscr{F} \{ t^3 \}
$$

Aplicando la propiedad de **diferenciación en frecuencia**:

$$
\mathscr{F} \{ t^n x(t) \} = j^n \cdot \frac{d^n}{d\omega^n} X(\omega)
$$

Con $n = 3$, y sabiendo que $x(t) = 1 \Rightarrow X(\omega) = \mathscr{F} \{ 1 \} = 2\pi \delta(\omega)$:

$$
\mathscr{F} \{ t^3 \} = j^3 \cdot \frac{d^3}{d\omega^3} \left[ 2\pi \delta(\omega) \right]
= j^3 \cdot 2\pi \cdot \frac{d^3}{d\omega^3} \delta(\omega)
$$

Sabemos que:

- $j^3 = j \cdot j \cdot j = -j$

Entonces:

$$
\mathscr{F} \{ t^3 \} = -j \cdot 2\pi \cdot \delta^{(3)}(\omega)
$$

Y por lo tanto:

$$
\mathscr{F} \{ 3 t^3 \} = 3 \cdot \left( -j \cdot 2\pi \cdot \delta^{(3)}(\omega) \right)
= -6j \pi \cdot \delta^{(3)}(\omega)
$$

---

**e).**

$$
X(\omega) = B \cdot \sum_{n = -\infty}^{\infty} \left( \frac{1}{a^2 + (\omega - n\omega_0)^2} + \frac{1}{a + j(\omega - n\omega_0)} \right),
\quad \omega_0 = \frac{2\pi}{T}, \quad B, T \in \mathbb{R}^+
$$

Donde: $n \in \{ 0, \pm1, \pm2, \dots \}$


Multiplicación de una señal aperiódica $f(t)$ por un tren de impulsos periódico:

$$
f(t) \cdot \sum_{k = -\infty}^{\infty} \delta(t - kT) \xrightarrow{\mathscr{F}} \frac{1}{T} \sum_{n = -\infty}^{\infty} F(\omega - n\omega_0)
$$


La constante de escala es $B$.

La función de forma base en frecuencia es:

$$
F(\omega) = \frac{1}{a^2 + \omega^2} + \frac{1}{a + j\omega}
$$

Entonces:

$$
f(t) = \mathscr{F}^{-1} \{ F(\omega) \} = \mathscr{F}^{-1} \left\{ \frac{1}{a^2 + \omega^2} \right\} + \mathscr{F}^{-1} \left\{ \frac{1}{a + j\omega} \right\}
$$



Para el primer término, aplicando la propiedad par de transformada:

$$
\mathscr{F}^{-1} \left\{ \frac{2a}{a^2 + \omega^2} \right\} = e^{-a|t|} \Rightarrow
\mathscr{F}^{-1} \left\{ \frac{1}{a^2 + \omega^2} \right\} = \frac{1}{2a} e^{-a|t|}
$$



Para el segundo término, aplicando la misma propiedad anterior:

$$
\mathscr{F}^{-1} \left\{ \frac{1}{a + j\omega} \right\} = e^{-at} u(t)
$$

Entonces la función base $f(t)$ es:

$$
f(t) = \frac{1}{2a} e^{-a|t|} + e^{-a t} u(t)
$$


Ahora se construye la señal final en el dominio del tiempo:

$$
x(t) = B \cdot f(t) \cdot \left( \sum_{k=-\infty}^{\infty} \delta(t - kT) \right)
$$

Reemplazando $f(t)$:

$$
x(t) = B \cdot \left( \frac{1}{2a} e^{-a|t|} + e^{-a t} u(t) \right) \cdot \sum_{k=-\infty}^{\infty} \delta(t - kT)
$$

Usando la propiedad de muestreo con deltas:

$$
x(t) = B \sum_{k = -\infty}^{\infty} f(kT) \cdot \delta(t - kT), \quad \text{donde} \quad f(kT) = \frac{1}{2a} e^{-a|kT|} + e^{-a kT} u(kT)
$$


**Entonces, la expresión final para la señal en el tiempo es:**

$$
x(t) = B \sum_{k = -\infty}^{\infty} \left( \frac{1}{2a} e^{-a|kT|} + e^{-a kT} u(kT) \right) \cdot \delta(t - kT)
$$

### ¿Qué es la modulación AM por detección coherente?

La modulación en amplitud (AM) por **detección coherente** consiste en recuperar la señal mensaje multiplicando la señal modulada por una réplica sincronizada de la portadora original (misma frecuencia y fase). Es decir, se requiere conocimiento exacto de la frecuencia portadora en el receptor.

#### Aplicaciones:
- Radiodifusión AM analógica.
- Transmisión de señales de voz o música.
- Comunicaciones analógicas punto a punto.
- Sensores de desplazamiento y telemetría.

Este tipo de detección permite recuperar con precisión la envolvente (señal mensaje), pero es más sensible a errores de sincronización que la detección no coherente.


```python
# Parámetros generales
fs = 1000       # Frecuencia de muestreo
T = 1.0         # Duración de la señal
t = np.linspace(0, T, int(fs*T), endpoint=False)

# Definimos el índice de modulación (ajustable por el usuario)
Im = 0.8

# ---------------------------
# 1. Señal mensaje: pulso rectangular
# ---------------------------
fm = 10
m_pulso = 1 * ((t > 0.3) & (t < 0.7)).astype(float)

# Portadora
fc = 100
Ac_pulso = np.max(np.abs(m_pulso)) / Im
c_pulso = Ac_pulso * np.cos(2 * np.pi * fc * t)

# Señal modulada
s_pulso = (1 + m_pulso / Ac_pulso) * c_pulso

# Detección coherente: multiplicar por portadora sincronizada
demod_pulso = s_pulso * c_pulso

# ---------------------------
# 2. Señal mensaje: coseno
# ---------------------------
m_cos = np.cos(2 * np.pi * fm * t)
Ac_cos = np.max(np.abs(m_cos)) / Im
c_cos = Ac_cos * np.cos(2 * np.pi * fc * t)
s_cos = (1 + m_cos / Ac_cos) * c_cos
demod_cos = s_cos * c_cos

# ---------------------------
# Gráficas en el tiempo
# ---------------------------
plt.figure(figsize=(14, 6))
plt.subplot(2, 2, 1)
plt.plot(t, m_pulso, label="Mensaje (pulso)")
plt.plot(t, s_pulso, label="Señal AM")
plt.plot(t, demod_pulso, label="Detectada coherente")
plt.title("Pulso Rectangular - Tiempo")
plt.legend()
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(t, m_cos, label="Mensaje (coseno)")
plt.plot(t, s_cos, label="Señal AM")
plt.plot(t, demod_cos, label="Detectada coherente")
plt.title("Coseno - Tiempo")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# ---------------------------
# Gráficas en frecuencia
# ---------------------------
freqs = rfftfreq(len(t), 1/fs)

S_pulso = rfft(s_pulso)
D_pulso = rfft(demod_pulso)
S_cos = rfft(s_cos)
D_cos = rfft(demod_cos)

plt.figure(figsize=(14, 6))
plt.subplot(2, 2, 1)
plt.plot(freqs, np.abs(S_pulso), label="AM (pulso)")
plt.plot(freqs, np.abs(D_pulso), label="Demodulada")
plt.title("Pulso Rectangular - Frecuencia")
plt.xlim(0, 200)
plt.grid()
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(freqs, np.abs(S_cos), label="AM (coseno)")
plt.plot(freqs, np.abs(D_cos), label="Demodulada")
plt.title("Coseno - Frecuencia")
plt.xlim(0, 200)
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

###  ¿Qué es la Distorsión Total Armónica (THD)?
La **distorsión total armónica (THD)** es una medida que indica cuántos armónicos (frecuencias múltiples no deseadas) están presentes en una señal además de su frecuencia principal.

- Una onda seno pura tiene solo una frecuencia: la fundamental.
- Cuando una señal se distorsiona (como al pasar por un rectificador), aparecen armónicos.
- El THD mide cuánta energía hay en esos armónicos respecto a la energía de la frecuencia fundamental:

$$
\text{THD} = \frac{\sqrt{V_2^2 + V_3^2 + \cdots + V_n^2}}{V_1}
$$

Donde $( V_1 )$ es la magnitud de la frecuencia fundamental y $( V_2, V_3, \dots )$ son los armónicos.

---

### ¿Qué es el Factor de Potencia por distorsión?

El **factor de potencia (FP)** representa la eficiencia en el uso de la energía eléctrica. Cuando hay distorsión por armónicos, se puede estimar así:

$$
\text{FP} = \frac{1}{\sqrt{1 + \text{THD}^2}}
$$

- Si el THD es bajo, el FP se acerca a 1 (condición ideal).
- Si el THD es alto, el FP disminuye (mala calidad de la señal).

---

### ¿Cómo se calcula el THD usando la FFT?

La **Transformada Rápida de Fourier (FFT)** descompone una señal en sus componentes de frecuencia. Para calcular el THD:

1. Se aplica FFT a la señal.
2. Se identifica la frecuencia fundamental (por ejemplo, 60 Hz).
3. Se suman las potencias de los armónicos (120 Hz, 180 Hz, etc.).
4. Se aplica la fórmula del THD mostrada arriba.

---

###  ¿Qué se analiza en las cargas resistiva y RC?

- **Carga resistiva pura**: la señal rectificada tiene muchos armónicos → alto THD.
- **Carga RC (resistencia + capacitor)**: el condensador actúa como filtro → reduce los armónicos → baja el THD.
- Cambiar **valores de R y C** permite observar cómo se afectan THD y FP.

---

###  Conclusión

- A mayor filtrado (capacitor grande), menor distorsión (bajo THD) y mejor factor de potencia.
- El análisis con FFT permite cuantificar de manera precisa la calidad de una señal y su eficiencia energética.


```python
# 🔧 Importar librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
from scipy.signal import lfilter

# ⚙️ Parámetros base
f = 60  # Frecuencia fundamental (Hz)
T = 1 / f
fs = 10000  # Frecuencia de muestreo (Hz)
t = np.linspace(0, 2 * T, int(2 * T * fs), endpoint=False)

# 🧮 Señal de entrada: seno y onda completa rectificada
vin = np.sin(2 * np.pi * f * t)
vrect = np.abs(vin)

# 📊 Función para calcular el THD y el Factor de Potencia
def compute_thd_and_pf(signal, fs, f0):
    Y = np.abs(rfft(signal))
    freqs = rfftfreq(len(signal), 1 / fs)
    idx_fund = np.argmin(np.abs(freqs - f0))
    V1 = Y[idx_fund]
    # Ensure V1 is not zero to avoid division by zero
    if V1 == 0:
        return np.nan, np.nan, freqs, Y  # Return NaN if fundamental is zero, also return freqs and Y

    # Calculate harmonic power excluding the fundamental
    # Sum of squares of all components minus square of fundamental
    total_power_spectrum = np.sum(Y**2) # Total power in the positive frequency spectrum
    fundamental_power = V1**2
    harmonic_power = total_power_spectrum - fundamental_power

    # Ensure harmonic power is not negative due to floating point inaccuracies
    thd = np.sqrt(max(0, harmonic_power)) / V1
    pf = 1 / np.sqrt(1 + thd**2)
    return thd, pf, freqs, Y # Return THD, PF, frequencies, and spectrum magnitude

# 🧪 Caso 1: Carga resistiva pura
print(f"🔴 Carga resistiva pura:")
thd_resistiva, pf_resistiva, freqs_R, Y_R = compute_thd_and_pf(vrect, fs, f)
print(f"   THD: {thd_resistiva*100:.2f} %")
print(f"   Factor de Potencia (Distorsión): {pf_resistiva:.3f}")

# 📊 Graficar señal rectificada (carga resistiva)
plt.figure(figsize=(10, 4))
plt.plot(t[:int(1.5*T*fs)], vrect[:int(1.5*T*fs)]) # Show about 1.5 periods
plt.title("Señal rectificada de onda completa (Carga Resistiva)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Voltaje (V)")
plt.grid(True)
plt.show()

# 🔎 Espectro de señal rectificada (carga resistiva)
plt.figure(figsize=(10, 4))
plt.plot(freqs_R, Y_R)
plt.title("Espectro de la señal rectificada (FFT)")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.xlim(0, f * 10) # Show up to 10th harmonic
plt.grid(True)
plt.show()


# 🧪 Caso 2: Carga RC en serie (R = 100 ohm, C variable)
R = 100
print("\n🟢 Carga RC en serie:")

# Store results for plotting comparison later
thds_rc = []
pfs_rc = []
C_vals_µF = []
vout_last = None
C_last = None
freqs_RC_last = None
Y_RC_last = None


for C in [0.000047, 0.0001, 0.00022, 0.00047]: # Test with typical capacitor values (uF converted to F)
    RC_time_constant = R * C
    # Simple digital filter approximation for RC low-pass
    # Using bilinear transform approach for better frequency response approximation
    # H(s) = 1 / (1 + sRC)
    # H(z) = (RC*fs + 1) / (RC*fs + 1 + (RC*fs - 1)*z^-1)  -- This is for H(s) = 1/(RC*s + 1)
    # Let's use a simpler form based on alpha = dt / (RC + dt) where dt = 1/fs
    alpha = 1 / (RC_time_constant * fs + 1)
    b = [alpha]
    a = [1, -(1-alpha)]

    vout = lfilter(b, a, vrect)
    thd_rc, pf_rc, freqs_RC, Y_RC = compute_thd_and_pf(vout, fs, f)

    thds_rc.append(thd_rc)
    pfs_rc.append(pf_rc)
    C_vals_µF.append(C * 1e6)

    print(f"   R = {R} Ω, C = {C*1e6:.0f} µF:")
    print(f"     THD: {thd_rc*100:.2f} %")
    print(f"     Factor de Potencia (Distorsión): {pf_rc:.3f}")

    # Store the last output for plotting
    vout_last = vout
    C_last = C * 1e6
    freqs_RC_last = freqs_RC
    Y_RC_last = Y_RC


# 📈 Comparación temporal de la última salida RC con la entrada rectificada
if vout_last is not None:
    plt.figure(figsize=(10, 4))
    plt.plot(t[:int(1.5*T*fs)], vrect[:int(1.5*T*fs)], label='Entrada rectificada')
    plt.plot(t[:int(1.5*T*fs)], vout_last[:int(1.5*T*fs)], label=f'Salida filtrada RC (C={C_last:.0f} µF)')
    plt.title(f'Respuesta RC - C = {C_last:.0f} µF (Tiempo)')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Voltaje (V)')
    plt.legend()
    plt.grid()
    plt.show()

# 🔎 Comparación espectral de la última salida RC con la entrada rectificada (resistiva)
if Y_RC_last is not None and Y_R is not None:
    plt.figure(figsize=(10, 4))
    plt.plot(freqs_R, Y_R, label="Carga Resistiva")
    plt.plot(freqs_RC_last, Y_RC_last, label=f"Carga RC (C={C_last:.0f} µF)")
    plt.title("Comparación Espectral (FFT)")
    plt.xlabel("Frecuencia (Hz)")
    plt.ylabel("Magnitud")
    plt.xlim(0, f * 10) # Show up to 10th harmonic
    plt.legend()
    plt.grid()
    plt.show()

# 📊 Graficar THD y FP vs C
if C_vals_µF and thds_rc and pfs_rc:
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(C_vals_µF, thds_rc, marker='o')
    plt.title("THD vs Capacitancia (R=100Ω)")
    plt.xlabel("Capacitancia (µF)")
    plt.ylabel("THD")
    plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.plot(C_vals_µF, pfs_rc, marker='o', color='orange')
    plt.title("Factor de Potencia vs Capacitancia (R=100Ω)")
    plt.xlabel("Capacitancia (µF)")
    plt.ylabel("Factor de Potencia")
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```
