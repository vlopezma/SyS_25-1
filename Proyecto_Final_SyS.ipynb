```python
%%writefile 📻_Concepto_Teorico.py

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import hilbert, firwin, lfilter
from scipy.fft import fft, fftfreq

# Configuración inicial del dashboard
st.set_page_config(page_title="Resumen Conceptos Clave", layout="wide")
st.title("📡 Resumen de Conceptos Clave: De Fourier al WiFi/5G")

st.markdown("""
Este dashboard interactivo presenta un resumen académico de los principales conceptos usados en la construcción de sistemas de comunicación digital modernos como WiFi y 5G.
Cada sección incluye teoría, visualizaciones y simulaciones en Python.
""")

# Secciones del Dashboard
secciones = [
    "Transformada de Fourier (FT/DFT/FFT)",
    "Filtrado Digital (FIR/IIR)",
    "Transformada de Hilbert y Señales Analíticas",
    "Modulación QAM e I/Q",
    "Multiplexación OFDM",
    "WiFi y 5G: Comunicación Inalámbrica Moderna"
]

opcion = st.sidebar.radio("Selecciona una sección:", secciones)

# ===============================
# Sección 1: Fourier
# ===============================
if opcion == secciones[0]:
    st.header("🎼 Transformada de Fourier (FT/DFT/FFT)")
    st.markdown("""
La Transformada de Fourier permite representar una señal en términos de sus componentes de frecuencia.

- **FT (Transformada Continua de Fourier)**: análisis en señales analógicas.
- **DFT (Transformada Discreta de Fourier)**: para secuencias digitales.
- **FFT (Fast Fourier Transform)**: algoritmo eficiente para calcular la DFT.

**Fórmula (DFT):**
$$
X[k] = \\sum_{n=0}^{N-1} x[n] \\cdot e^{-j2\\pi kn/N}
$$
""")

    # Simulación
    f1 = st.slider("Frecuencia f1 (Hz)", 1, 50, 5)
    f2 = st.slider("Frecuencia f2 (Hz)", 1, 50, 15)
    fs = 200
    t = np.arange(0, 1, 1/fs)
    x = np.sin(2*np.pi*f1*t) + 0.5*np.sin(2*np.pi*f2*t)
    X = np.abs(fft(x))
    freqs = fftfreq(len(t), 1/fs)

    fig, ax = plt.subplots(2, 1, figsize=(10, 4))
    ax[0].plot(t, x)
    ax[0].set_title("Señal en el tiempo")
    ax[1].stem(freqs[:len(freqs)//2], X[:len(X)//2], basefmt=" ")
    ax[1].set_title("Magnitud del espectro")
    st.pyplot(fig)

# ===============================
# Sección 2: Filtrado Digital
# ===============================
elif opcion == secciones[1]:
    st.header("🔎 Filtrado Digital (FIR/IIR)")
    st.markdown("""
Los filtros digitales permiten modificar el contenido espectral de una señal.

- **FIR (Finite Impulse Response)**: respuesta finita.
- **IIR (Infinite Impulse Response)**: respuesta infinita, suelen ser más eficientes.

**Ejemplo de FIR (Filtro Paso-Bajo):**
""")

    fc = st.slider("Frecuencia de corte (Hz)", 10, 90, 30)
    fs = 200
    t = np.arange(0, 1, 1/fs)
    x = np.sin(2*np.pi*5*t) + 0.5*np.sin(2*np.pi*15*t)
    nyq = fs / 2
    taps = firwin(101, fc/nyq)
    y = lfilter(taps, 1.0, x)

    fig, ax = plt.subplots(2, 1, figsize=(10, 4))
    ax[0].plot(t, x, label="Original")
    ax[0].plot(t, y, label="Filtrada")
    ax[0].legend()
    ax[0].set_title("Señales antes y después del filtrado")
    ax[1].plot(np.abs(fft(y))[:len(y)//2], label="Espectro filtrado")
    ax[1].legend()
    st.pyplot(fig)

# ===============================
# Sección 3: Hilbert
# ===============================
elif opcion == secciones[2]:
    st.header("📐 Transformada de Hilbert y Señales Analíticas")
    st.markdown(r"""
La Transformada de Hilbert permite obtener la representación analítica de una señal.

**Definición:**
$$
\hat{x}(t) = \mathcal{H}\{x(t)\} = \frac{1}{\pi} \text{p.v.} \int_{-\infty}^{\infty} \frac{x(\tau)}{t - \tau} d\tau
$$

La señal analítica se define como:
$$
x_a(t) = x(t) + j \hat{x}(t)
$$
""")

    fs = 200
    t = np.arange(0, 1, 1/fs)
    x = np.sin(2*np.pi*5*t)
    analytic = hilbert(x)
    I = np.real(analytic)
    Q = np.imag(analytic)

    fig, ax = plt.subplots(2, 1, figsize=(10, 4))
    ax[0].plot(t, I, label="I(t)")
    ax[0].plot(t, Q, label="Q(t)")
    ax[0].legend()
    ax[0].set_title("Componentes en Fase y Cuadratura")
    ax[1].plot(I, Q)
    ax[1].set_title("Plano Complejo: I vs Q")
    st.pyplot(fig)

# ===============================
# Sección 4: QAM
# ===============================
elif opcion == secciones[3]:
    st.header("📡 Modulación QAM e I/Q")
    st.markdown("""
QAM (Quadrature Amplitude Modulation) combina dos señales: una en fase (I) y otra en cuadratura (Q), permitiendo transmitir más bits por símbolo.

**16-QAM:**
- Se utilizan 4 niveles para I y 4 niveles para Q.
- Se pueden representar 16 símbolos únicos en el plano I/Q.

**Constelación de 16-QAM:**
""")
    qam_order = 16
    M = int(np.sqrt(qam_order))
    symbols = np.array([complex(i, q) for i in range(-M+1, M, 2) for q in range(-M+1, M, 2)])
    np.random.seed(0)
    data = np.random.choice(symbols, size=100)

    fig, ax = plt.subplots()
    ax.scatter(data.real, data.imag, c='blue')
    ax.set_title("Diagrama de Constelación - 16QAM")
    ax.set_xlabel("I")
    ax.set_ylabel("Q")
    st.pyplot(fig)

# ===============================
# Sección 5: OFDM
# ===============================
elif opcion == secciones[4]:
    st.header("🔀 Multiplexación por División de Frecuencia Ortogonal (OFDM)")
    st.markdown("""
OFDM divide el ancho de banda disponible en múltiples subportadoras ortogonales.

- Cada subportadora modula una porción de la señal (p. ej., con QAM).
- Se utiliza la IFFT para pasar del dominio de la frecuencia al tiempo.

**Ventajas:**
- Alta eficiencia espectral.
- Robusta contra desvanecimiento selectivo.
""")

# ===============================
# Sección 6: WiFi y 5G
# ===============================
elif opcion == secciones[5]:
    st.header("📶 Comunicación Inalámbrica: WiFi y 5G")
    st.markdown("""
Las tecnologías WiFi y 5G utilizan QAM y OFDM para enviar datos de forma eficiente.

**Enlace con conceptos:**
- **WiFi/5G** → usan **QAM** sobre subportadoras OFDM.
- **I/Q** → representan los datos modulados digitalmente.
- **Hilbert** → para construir señales analíticas.
- **FFT** → para el análisis y síntesis espectral.
""")
    st.markdown("Todo esto converge en sistemas modernos de comunicación digital que permiten la hiperconectividad actual.")
```

```python
%%writefile 🎵_Entrada_Senal.py

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter, freqz
from scipy.signal.windows import hann
from numpy.fft import fft, fftfreq

st.set_page_config(page_title="Análisis Señal", page_icon="🔍")

st.title("🔍 Análisis de Señales Sintéticas y Filtrado en el Dominio de la Frecuencia")

st.markdown("""
Este módulo simula y analiza señales sintéticas compuestas por varias frecuencias sinusoidales. A través de una interfaz interactiva, se permite:
- Construir señales personalizadas.
- Visualizar su comportamiento en el tiempo.
- Estudiar su contenido espectral mediante FFT.
- Aplicar un **filtro digital paso-bajo** y observar sus efectos.
- Explorar el **diagrama de Bode** del filtro utilizado.

---

""")

# --- Parámetros de señal (sidebar) ---
st.sidebar.header("⚙️ Parámetros de la señal")
fs = st.sidebar.slider("Frecuencia de muestreo [Hz]", 500, 5000, 1000, 100)
dur = st.sidebar.slider("Duración [s]", 1.0, 5.0, 2.0, 0.5)

f1 = st.sidebar.slider("Frecuencia componente 1", 5, fs//2 - 1, 20, 1)
f2 = st.sidebar.slider("Frecuencia componente 2", 5, fs//2 - 1, 80, 1)
f3 = st.sidebar.slider("Frecuencia componente 3", 5, fs//2 - 1, 200, 1)

# --- Construcción de la señal ---
t = np.linspace(0, dur, int(fs*dur), endpoint=False)
x = (np.sin(2*np.pi*f1*t) +
     0.7 * np.sin(2*np.pi*f2*t) +
     0.5 * np.sin(2*np.pi*f3*t))

st.markdown("### ⏱️ Visualización en el tiempo")
st.markdown(f"Generamos una señal compuesta por tres frecuencias: `{f1} Hz`, `{f2} Hz` y `{f3} Hz`. La duración es de `{dur}` segundos a una tasa de muestreo de `{fs} Hz`.")
fig1, ax1 = plt.subplots()
ax1.plot(t, x)
ax1.set_xlabel("Tiempo [s]")
ax1.set_ylabel("Amplitud")
ax1.set_title("Señal Sintética en el Tiempo")
st.pyplot(fig1)

# --- Aplicación de ventana + FFT ---
st.markdown("### 📈 Espectro (FFT) de la Señal")
st.markdown("""
Para observar las componentes espectrales, se aplica la **Transformada Rápida de Fourier (FFT)**.
Se utiliza una ventana de Hann para minimizar fugas espectrales.
""")
ventana = hann(len(x))
x_win = x * ventana
X_fft = fft(x_win)
f_fft = fftfreq(len(x), 1/fs)
mag_fft = 20 * np.log10(np.abs(X_fft) + 1e-6)

fig2, ax2 = plt.subplots()
ax2.plot(f_fft[:len(f_fft)//2], mag_fft[:len(mag_fft)//2])
ax2.set_xlabel("Frecuencia [Hz]")
ax2.set_ylabel("Magnitud [dB]")
ax2.set_title("Contenido Espectral de la Señal (FFT)")
ax2.grid()
st.pyplot(fig2)

# --- Parámetros del filtro paso-bajo ---
st.sidebar.header("🧹 Filtro Paso-Bajo")
fc = st.sidebar.slider("Frecuencia de corte [Hz]", 10, fs//2 - 10, 100, 10)
orden = st.sidebar.slider("Orden del filtro", 1, 10, 4)

# --- Diseño del filtro Butterworth ---
def butter_lowpass(cutoff, fs, order=4):
    nyq = 0.5 * fs
    norm_cutoff = cutoff / nyq
    b, a = butter(order, norm_cutoff, btype='low', analog=False)
    return b, a

b, a = butter_lowpass(fc, fs, orden)
y = lfilter(b, a, x)

# --- Señal filtrada vs original ---
st.markdown("### 🧼 Señal Filtrada")
st.markdown(f"""
Se aplicó un filtro **Butterworth paso-bajo** de orden `{orden}` y frecuencia de corte `{fc} Hz`.
Este filtro elimina frecuencias por encima de ese umbral, conservando las más bajas.
""")
fig3, ax3 = plt.subplots()
ax3.plot(t, x, label="Original", alpha=0.5, linestyle="--")
ax3.plot(t, y, label="Filtrada", linewidth=2)
ax3.set_title("Comparación en el Dominio del Tiempo")
ax3.set_xlabel("Tiempo [s]")
ax3.set_ylabel("Amplitud")
ax3.legend()
ax3.grid()
st.pyplot(fig3)

# --- FFT filtrada ---
Y_fft = fft(y * hann(len(y)))
mag_Y = 20 * np.log10(np.abs(Y_fft) + 1e-6)

st.markdown("### 📉 Espectro después del Filtrado")
st.markdown("Se puede observar que las componentes de alta frecuencia han sido atenuadas con éxito.")
fig4, ax4 = plt.subplots()
ax4.plot(f_fft[:len(f_fft)//2], mag_Y[:len(mag_Y)//2])
ax4.set_xlabel("Frecuencia [Hz]")
ax4.set_ylabel("Magnitud [dB]")
ax4.set_title("FFT de la Señal Filtrada")
ax4.grid()
st.pyplot(fig4)

# --- Diagrama de Bode del filtro ---
st.markdown("### 📊 Diagrama de Bode del Filtro")
st.markdown("El diagrama de Bode muestra cómo el filtro atenúa las frecuencias por encima del umbral de corte.")

w, h = freqz(b, a, worN=8000)
frecs = 0.5 * fs * w / np.pi
bode = 20 * np.log10(abs(h))

fig5, ax5 = plt.subplots()
ax5.semilogx(frecs, bode)
ax5.set_xlabel("Frecuencia [Hz]")
ax5.set_ylabel("Ganancia [dB]")
ax5.set_title("Respuesta en Frecuencia (Bode) del Filtro")
ax5.grid(which='both', linestyle='--', linewidth=0.5)
ax5.axvline(fc, color='red', linestyle=':', label="fc = {} Hz".format(fc))
ax5.legend()
st.pyplot(fig5)

st.markdown("---")
st.success("✅ ¡Simulación completa! Ya puedes cambiar parámetros y observar sus efectos en tiempo real.")
```

```python
%%writefile 🔁_Senales_IQ.py

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert, butter, freqz, lfilter
from scipy.signal.windows import hann
from scipy.fft import fft, fftfreq, fftshift
from scipy.io import wavfile
import subprocess
import yt_dlp as youtube_dl

st.set_page_config(page_title="🔁 Señales I/Q", page_icon="🌀", layout="wide")
st.title("🌀 Construcción de Señales I/Q usando Transformada de Hilbert")

st.markdown("""
## 🎯 Objetivo

Analizar una señal real de audio en su forma **analítica** usando la **Transformada de Hilbert**, y estudiar sus componentes en **fase ($I(t)$)** y **cuadratura ($Q(t)$)**.
Estas componentes son fundamentales para sistemas de comunicación digital como **WiFi** y **5G**, donde se usan en modulaciones como QAM.

---

## 🧠 Fundamento

Dada una señal real $x(t)$, su versión analítica se define como:

$$
x_a(t) = x(t) + j \cdot \mathcal{H}\{x(t)\} = I(t) + jQ(t)
$$

Donde:
- $I(t) = x(t)$: componente en fase.
- $Q(t) = \mathcal{H}\{x(t)\}$: componente en cuadratura (defasada 90°).

Esta forma permite representar la señal como un número complejo en el **plano I/Q**, clave en comunicación digital.

---
""")

# ==== Parámetros ====
st.sidebar.header("⚙️ Parámetros de análisis")
fs = st.sidebar.slider("Frecuencia de muestreo (Hz)", 8000, 44100, 16000, 1000)
dur_max = st.sidebar.slider("Duración a analizar (s)", 1, 10, 3, 1)
url = st.sidebar.text_input("🔗 Enlace de YouTube para extraer audio:")

# ==== Descarga de audio ====
audio_loaded = False
if url:
    try:
        with st.spinner("🎵 Descargando audio desde YouTube..."):
            ydl_opts = {'format': 'bestaudio', 'outtmpl': 'temp_audio.%(ext)s'}
            with youtube_dl.YoutubeDL(ydl_opts) as ydl:
                ydl.download([url])

            subprocess.run(['ffmpeg', '-y', '-i', 'temp_audio.webm',
                            '-ar', str(fs), '-ac', '1', 'audio.wav'],
                           stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

            sr, audio = wavfile.read('audio.wav')
            audio = audio.astype(np.float32)
            audio /= np.max(np.abs(audio))
            audio = audio[:int(fs * dur_max)]
            t = np.linspace(0, len(audio)/fs, len(audio), endpoint=False)
            audio_loaded = True
    except Exception as e:
        st.error(f"❌ Error procesando audio: {e}")

# ==== Procesamiento I/Q ====
if audio_loaded:
    st.markdown("## 🧪 Componentes $I(t)$ y $Q(t)$")

    Q = np.imag(hilbert(audio))

    fig1, ax1 = plt.subplots(2, 1, figsize=(10, 4), sharex=True)
    ax1[0].plot(t, audio, label="I(t)")
    ax1[1].plot(t, Q, color='orange', label="Q(t)")
    ax1[0].set_title("🟦 Componente en Fase: $I(t)$")
    ax1[1].set_title("🟧 Componente en Cuadratura: $Q(t)$")
    ax1[0].legend(); ax1[1].legend()
    ax1[0].grid(); ax1[1].grid()
    st.pyplot(fig1)

    st.markdown("""
### 📌 Interpretación

La Transformada de Hilbert genera una señal $Q(t)$ que está **defasada 90°** respecto a $I(t)$.
Este par ortogonal es la base para representar modulaciones complejas como QAM.

---
""")

    st.markdown("## 🧿 Plano Complejo I/Q")
    st.markdown("""
### 📌 ¿Qué muestra esta gráfica?

Cada punto representa un instante de la señal como un número complejo $x_a(t) = I(t) + jQ(t)$.
En modulaciones como 16-QAM, estos puntos forman constelaciones discretas.
Aquí, al ser una señal de audio, se obtiene una **nube continua**.

---
""")
    fig2, ax2 = plt.subplots(figsize=(6, 4))
    ax2.scatter(audio, Q, s=1, alpha=0.5)
    ax2.set_xlabel("$I(t)$"); ax2.set_ylabel("$Q(t)$")
    ax2.set_title("Plano complejo I/Q"); ax2.grid()
    st.pyplot(fig2)

    st.markdown("## 📈 Espectro de $I(t)$ y $Q(t)$ con ventana de Hann")
    st.markdown("""
### 📌 ¿Por qué usar ventana de Hann?

Al aplicar la FFT, la señal no es periódica dentro del intervalo analizado.
La ventana de Hann suaviza los extremos de la señal para evitar **fugas espectrales**, mejorando la precisión del análisis.

---
""")
    N = len(audio)
    window = hann(N)
    audio_win = audio * window
    Q_win = Q * window

    f = fftshift(fftfreq(N, 1/fs))
    fft_i = fftshift(np.abs(fft(audio_win)) / np.sum(window))
    fft_q = fftshift(np.abs(fft(Q_win)) / np.sum(window))

    fig3, ax3 = plt.subplots(figsize=(10, 3))
    ax3.plot(f, fft_i, label="|FFT(I)|")
    ax3.plot(f, fft_q, label="|FFT(Q)|", linestyle='--')
    ax3.set_xlim(0, fs//2)
    ax3.set_xlabel("Frecuencia [Hz]")
    ax3.set_ylabel("Magnitud")
    ax3.set_title("Contenido espectral (ventana de Hann)")
    ax3.legend(); ax3.grid()
    st.pyplot(fig3)

    st.markdown("## 📉 Diagrama de Bode del filtro paso bajo")

    b, a = butter(4, 0.3)
    w, h = freqz(b, a, worN=8000)
    phase_unwrapped = np.unwrap(np.angle(h))

    fig_bode, (ax_bode_mag, ax_bode_phase) = plt.subplots(2, 1, figsize=(8, 4))
    ax_bode_mag.plot(fs * w / (2*np.pi), 20*np.log10(np.abs(h)))
    ax_bode_mag.set_title("Magnitud [dB]"); ax_bode_mag.set_ylabel("dB"); ax_bode_mag.grid()

    ax_bode_phase.plot(fs * w / (2*np.pi), phase_unwrapped * 180/np.pi)
    ax_bode_phase.set_title("Fase [°] (desenrollada)")
    ax_bode_phase.set_xlabel("Frecuencia [Hz]"); ax_bode_phase.set_ylabel("Grados"); ax_bode_phase.grid()
    st.pyplot(fig_bode)

    st.markdown("""
### 📌 ¿Para qué sirve este filtro?

Los filtros IIR como Butterworth se usan en sistemas reales para:
- Eliminar ruido.
- Separar bandas laterales (en SSB, QAM).
- Definir canales de comunicación.

---
""")

    st.markdown("## 📊 Magnitud y Fase de $x_a(t)$")

    analytic = audio + 1j * Q
    magnitude = np.abs(analytic)
    phase = np.angle(analytic)
    mask = magnitude > 0.01

    fig_mag_phase, ax = plt.subplots(2, 1, figsize=(10, 4), sharex=True)
    ax[0].plot(t, magnitude, color='purple')
    ax[1].plot(t[mask], phase[mask], color='green')
    ax[0].set_title("🔺 Magnitud de $x_a(t)$")
    ax[1].set_title("🔻 Fase de $x_a(t)$")
    ax[0].grid(); ax[1].grid()
    st.pyplot(fig_mag_phase)

    st.markdown("""
### 📌 ¿Qué representa esta gráfica?

- La **magnitud** corresponde a la envolvente de la señal.
- La **fase** indica variaciones de frecuencia instantánea, útil en demodulación.

---
""")

    # === MODULACIÓN QAM REAL ===
    st.markdown("## 📡 Modulación QAM real con $I(t)$ y $Q(t)$")
    fc = st.sidebar.slider("🎚️ Frecuencia de portadora (Hz)", 1000, fs//2 - 500, 4000, 500)

    portadora_cos = np.cos(2 * np.pi * fc * t)
    portadora_sin = np.sin(2 * np.pi * fc * t)
    s_qam = audio * portadora_cos - Q * portadora_sin

    st.markdown("### ⏱️ Señal QAM modulada en el dominio del tiempo")
    fig4, ax4 = plt.subplots(figsize=(10, 3))
    ax4.plot(t[:1000], s_qam[:1000])
    ax4.set_title("Señal QAM Modulada (primeros mil puntos)")
    ax4.set_xlabel("Tiempo [s]")
    ax4.set_ylabel("Amplitud")
    ax4.grid()
    st.pyplot(fig4)

    st.markdown("### 📈 Espectro de la señal modulada")
    st.markdown("""
### 📌 ¿Qué se observa en este espectro?

Este espectro nos permite ver cómo la información de I y Q se ha desplazado a una frecuencia central (la portadora $f_c$), formando una señal de banda pasante que puede transmitirse.

---
""")
    s_qam_win = s_qam * hann(len(s_qam))
    fft_qam = fftshift(np.abs(fft(s_qam_win)) / np.sum(window))
    fig5, ax5 = plt.subplots(figsize=(10, 3))
    ax5.plot(f, fft_qam)
    ax5.set_title("Espectro de la señal QAM modulada")
    ax5.set_xlabel("Frecuencia [Hz]")
    ax5.set_ylabel("Magnitud")
    ax5.set_xlim(0, fs//2)
    ax5.grid()
    st.pyplot(fig5)

    # === CANAL CON RUIDO ===
    st.markdown("## 📡 Canal con ruido (AWGN)")
    snr_db = st.sidebar.slider("🔊 Relación señal-ruido (SNR) [dB]", 0, 60, 30, 5)
    signal_power = np.mean(s_qam**2)
    noise_power = signal_power / (10**(snr_db/10))
    ruido = np.random.normal(0, np.sqrt(noise_power), size=s_qam.shape)
    s_ruido = s_qam + ruido

    st.markdown("### 🔊 Señal recibida con ruido")
    fig6, ax6 = plt.subplots(figsize=(10, 3))
    ax6.plot(t[:1000], s_ruido[:1000])
    ax6.set_title("Señal QAM con ruido (primeros mil puntos)")
    ax6.set_xlabel("Tiempo [s]")
    ax6.set_ylabel("Amplitud")
    ax6.grid()
    st.pyplot(fig6)

    st.markdown("### 📉 Espectro de la señal con ruido")
    st.markdown("""
### 📌 ¿Qué se observa en este espectro?

Aquí observamos cómo el **ruido blanco gaussiano** afecta la señal transmitida. Verás que el espectro pierde definición, especialmente si el SNR es bajo.

---
""")
    s_ruido_win = s_ruido * hann(len(s_ruido))
    fft_ruido = fftshift(np.abs(fft(s_ruido_win)) / np.sum(window))
    fig7, ax7 = plt.subplots(figsize=(10, 3))
    ax7.plot(f, fft_ruido)
    ax7.set_title("Espectro de la señal recibida")
    ax7.set_xlabel("Frecuencia [Hz]")
    ax7.set_ylabel("Magnitud")
    ax7.set_xlim(0, fs//2)
    ax7.grid()
    st.pyplot(fig7)

    st.success("✅ ¡Señal transmitida y canal con ruido implementados!")

    # === DEMODULACIÓN ===
    st.markdown("## 🔄 Demodulación y recuperación de $I(t)$ y $Q(t)$")
    st.markdown("""
Multiplicamos la señal ruidosa por coseno y seno para traerla de vuelta a banda base y aplicamos un filtro pasa bajos para recuperar I(t) y Q(t).
""")

    # Reutilizamos la portadora para demodular
    I_demod = s_ruido * portadora_cos
    Q_demod = -s_ruido * portadora_sin

    # Filtro paso bajo para recuperar las señales en banda base
    b_lp, a_lp = butter(4, 0.3)
    I_rec = lfilter(b_lp, a_lp, I_demod)
    Q_rec = lfilter(b_lp, a_lp, Q_demod)

    st.markdown("### 🔁 Señales $I(t)$ y $Q(t)$ recuperadas")
    fig8, ax8 = plt.subplots(2, 1, figsize=(10, 4), sharex=True)
    ax8[0].plot(t, audio, label="I original", alpha=0.7)
    ax8[0].plot(t, I_rec, label="I recuperada", linestyle='--')
    ax8[0].set_title("🟦 Comparación de $I(t)$")
    ax8[0].legend(); ax8[0].grid()

    ax8[1].plot(t, Q, label="Q original", alpha=0.7)
    ax8[1].plot(t, Q_rec, label="Q recuperada", linestyle='--')
    ax8[1].set_title("🟧 Comparación de $Q(t)$")
    ax8[1].legend(); ax8[1].grid()

    st.pyplot(fig8)

    st.markdown("""
### 📌 ¿Qué se observa?

- Las señales recuperadas se asemejan a las originales, aunque pueden tener ruido y desfasajes menores.
- Esto demuestra que el sistema permite **transmitir y recuperar señales analógicas** usando modulación QAM.

---
""")

    st.success("✅ ¡Demodulación completada!")
```

```python
%%writefile 🧩_Sistema_Completo.py

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert, butter, lfilter
from scipy.signal.windows import hann
from scipy.fft import fft, fftfreq
from scipy.io import wavfile
import subprocess
import os

try:
    import yt_dlp as youtube_dl
except ImportError:
    youtube_dl = None

st.set_page_config(page_title="🧩 Sistema Completo QAM", layout="wide")
st.title("🧩 Sistema Completo QAM + Canal AWGN")

# Sidebar: Parámetros generales
st.sidebar.header("⚙️ Parámetros Generales")
fs = st.sidebar.slider("Frecuencia de muestreo [Hz]", 4000, 44100, 8000, 1000)
dur = st.sidebar.slider("Duración simulación [s]", 1.0, 10.0, 2.0, 0.5)
zoom_t = st.sidebar.slider("Duración visualizada (zoom) [ms]", 1, int(dur*1000), 10)

t = np.linspace(0, dur, int(fs * dur), endpoint=False)
x = np.zeros_like(t)

# Fuente de señal
modo = st.sidebar.radio("Fuente de señal", ["Sintética", "YouTube"])
if modo == "Sintética":
    st.markdown("### 📡 Fuente: Señal sintética")
    f1 = st.sidebar.slider("f1 [Hz]", 50, fs//2-1, 100)
    f2 = st.sidebar.slider("f2 [Hz]", 100, fs//2-1, 300)
    x = np.sin(2*np.pi*f1*t) + 0.7 * np.sin(2*np.pi*f2*t)
    st.info("✅ Señal sintética generada correctamente.")
else:
    st.markdown("### 🎧 Fuente: Audio desde YouTube")
    yt_url = st.sidebar.text_input("🔗 Enlace YouTube (solo audio):")
    if yt_url and youtube_dl:
        try:
            with st.spinner("Descargando y procesando audio..."):
                ydl_opts = {'format': 'bestaudio', 'outtmpl': 'temp_audio.%(ext)s'}
                with youtube_dl.YoutubeDL(ydl_opts) as ydl:
                    ydl.download([yt_url])
                subprocess.run(['ffmpeg', '-y', '-i', 'temp_audio.webm',
                                '-ar', str(fs), '-ac', '1', 'audio.wav'],
                               stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
                sr, audio = wavfile.read('audio.wav')
                audio = audio.astype(np.float32)
                audio /= np.max(np.abs(audio) + 1e-6)
                if len(audio) >= len(t):
                    x = audio[:len(t)]
                else:
                    x = np.pad(audio, (0, len(t) - len(audio)), mode='constant')
                st.success("✅ Audio cargado y procesado.")
        except Exception as e:
            st.error(f"❌ Error al cargar audio: {e}")
    else:
        st.warning("🔁 Ingresa un enlace válido o asegúrate de tener yt_dlp instalado.")

# Zoom
t_vis = t[:int((zoom_t/1000) * fs)]
x_vis = x[:int((zoom_t/1000) * fs)]

# Tiempo
st.markdown("### ⏱️ Señal en el dominio del tiempo")
figT, axT = plt.subplots()
axT.plot(t_vis, x_vis)
axT.set_xlabel("Tiempo [s]")
axT.set_ylabel("Amplitud")
axT.set_title(f"Señal (primeros {zoom_t} ms)")
axT.grid()
st.pyplot(figT)

# FFT
st.markdown("### 📊 Espectro (FFT)")
X = fft(x * hann(len(x)))
f = fftfreq(len(x), 1/fs)
figF, axF = plt.subplots()
axF.plot(f[:len(f)//2], 20*np.log10(np.abs(X[:len(X)//2]) + 1e-6))
axF.set_xlabel("Frecuencia [Hz]")
axF.set_ylabel("Magnitud [dB]")
axF.set_title("Espectro de la señal")
axF.grid()
st.pyplot(figF)

# Filtro paso bajo
st.header("🧹 Filtro paso bajo (Butterworth)")
fc = st.sidebar.slider("Frecuencia de corte [Hz]", 100, fs//2-100, 500)
b, a = butter(4, fc/(fs/2), btype='low')
x_f = lfilter(b, a, x)

figf, axf = plt.subplots()
axf.plot(t_vis, x_vis, label="Original", alpha=0.4)
axf.plot(t_vis, x_f[:len(t_vis)], label="Filtrada", linewidth=2)
axf.set_title("Señal original vs filtrada")
axf.set_xlabel("Tiempo [s]")
axf.set_ylabel("Amplitud")
axf.grid(); axf.legend()
st.pyplot(figf)

# Hilbert (I/Q)
st.header("📌 Componentes I/Q")
analytic = hilbert(x_f)
I = np.real(analytic)
Q = np.imag(analytic)
figIQ, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.plot(t_vis, I[:len(t_vis)])
ax2.plot(t_vis, Q[:len(t_vis)])
ax1.set_title("I(t) – en fase")
ax2.set_title("Q(t) – en cuadratura")
ax2.set_xlabel("Tiempo [s]")
ax1.grid(); ax2.grid()
st.pyplot(figIQ)

# Modulación QAM
st.header("📡 Modulación QAM")
M = st.sidebar.selectbox("Orden QAM (M)", [4, 16, 64], index=1)
k = int(np.log2(M))
sym_rate = st.sidebar.slider("Símbolos por segundo", 50, int(fs/2), 500)
n_sym = min(int(sym_rate * dur), len(t)//10)
bits = np.random.randint(0, 2, size=(n_sym, k))

def qam_mod(bits, M):
    if M == 4:
        return (2*bits[:,0]-1) + 1j*(2*bits[:,1]-1)
    elif M == 16:
        b0 = 2*bits[:,0] + bits[:,1]
        b1 = 2*bits[:,2] + bits[:,3]
        return (2*b0 - 3) + 1j*(2*b1 - 3)
    elif M == 64:
        b0 = 4*bits[:,0] + 2*bits[:,1] + bits[:,2]
        b1 = 4*bits[:,3] + 2*bits[:,4] + bits[:,5]
        return (2*b0 - 7) + 1j*(2*b1 - 7)
    else:
        raise ValueError("QAM orden no soportado")

symbols = qam_mod(bits, M)
samples_per_symbol = len(t) // n_sym
s_sym = np.repeat(symbols, samples_per_symbol)[:len(t)]

figSYM, axSYM = plt.subplots()
axSYM.scatter(symbols.real, symbols.imag, alpha=0.6, color='blue', label="Transmitidos")
axSYM.set_title(f"Constelación {M}-QAM transmitida")
axSYM.set_xlabel("I")
axSYM.set_ylabel("Q")
axSYM.grid(); axSYM.legend()
st.pyplot(figSYM)

# Portadora
fc_qam = st.sidebar.slider("Frecuencia portadora [Hz]", 500, fs//2-1000, 2000)
cos_c = np.cos(2*np.pi*fc_qam*t)
sin_c = np.sin(2*np.pi*fc_qam*t)
s_tx = np.real(s_sym)*cos_c - np.imag(s_sym)*sin_c

# Canal AWGN
st.header("📻 Canal AWGN")
snr = st.sidebar.slider("SNR [dB]", 0, 40, 20)
P_sig = np.mean(s_tx**2)
noise = np.random.normal(0, np.sqrt(P_sig / 10**(snr/10)), size=len(s_tx))
s_rx = s_tx + noise

figN, axN = plt.subplots()
axN.plot(t_vis, s_rx[:len(t_vis)])
axN.set_title("Señal recibida con ruido (zoom)")
axN.set_xlabel("Tiempo [s]")
axN.set_ylabel("Amplitud")
axN.grid()
st.pyplot(figN)

# Demodulación
st.header("🔄 Demodulación y Constelación recibida")
I_rx = s_rx * cos_c
Q_rx = -s_rx * sin_c
I_f = lfilter(b, a, I_rx)
Q_f = lfilter(b, a, Q_rx)
sym_rec = I_f[::samples_per_symbol][:n_sym] + 1j * Q_f[::samples_per_symbol][:n_sym]

figRC, axRC = plt.subplots()
axRC.scatter(symbols.real, symbols.imag, alpha=0.3, color='blue', label="Transmitidos")
axRC.scatter(sym_rec.real, sym_rec.imag, alpha=0.6, color='red', label="Recibidos")
axRC.set_title("Constelación: enviado (azul) vs recibido (rojo)")
axRC.set_xlabel("I")
axRC.set_ylabel("Q")
axRC.grid(); axRC.legend()
st.pyplot(figRC)

st.success("✅ Simulación finalizada correctamente.")
```
