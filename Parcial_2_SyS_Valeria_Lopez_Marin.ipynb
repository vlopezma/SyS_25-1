```python
%%writefile 🎯_Punto_1.py


import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import TransferFunction, bode, step, impulse, lti

# Función para simular retroalimentación negativa

def feedback(sys1, sys2=1):
    num1, den1 = sys1.num, sys1.den
    if isinstance(sys2, (int, float)):
        num2, den2 = [sys2], [1]
    else:
        num2, den2 = sys2.num, sys2.den

    num_fb = np.polymul(num1, num2)
    den_fb = np.polymul(den1, den2)
    den_total = np.polyadd(den_fb, num_fb)
    return TransferFunction(num1, den_total)

st.set_page_config(
    page_title="Punto 1",
    page_icon="🎯",
    layout="wide"
)

# Título y descripción
st.title("🎯 Punto 1: Función de Transferencia y Modelo Eléctrico Equivalente")

# Sección 1: Transformada de Laplace
st.markdown("## Transformada de Laplace")

st.markdown("""
Aplicando la **transformada unilateral de Laplace** con condiciones iniciales cero al sistema masa-resorte-amortiguador, obtenemos:
""")

st.latex(r"F_E(s) = (m s^2 + c s + k) \cdot Y(s)")

st.markdown("""
De aquí se despeja la **función de transferencia en lazo abierto**, que relaciona la salida \( Y(s) \) con la entrada \( F_E(s) \):
""")

st.latex(r"H(s) = \frac{Y(s)}{F_E(s)} = \frac{1}{m s^2 + c s + k}")

st.markdown("---")

# Sección 2: Modelo Eléctrico Equivalente
st.markdown("## ⚡ Modelo Eléctrico Equivalente")

st.markdown("""
Se considera un circuito **RLC serie** con entrada \( V_i(t) \) y salida \( V_o(t) \), donde la salida se toma sobre el **capacitor**.
El objetivo es encontrar un sistema eléctrico con comportamiento equivalente al sistema mecánico.

La ecuación del circuito en el dominio de Laplace es:
""")

st.latex(r"V_i(s) = \left( L C s^2 + \frac{L}{R} s + 1 \right) V_o(s)")

st.markdown("De aquí, la **función de transferencia** es:")
st.latex(r"\frac{V_o(s)}{V_i(s)} = \frac{1}{L C s^2 + \frac{L}{R} s + 1}")

st.markdown("---")

# Nota final
st.info("Ambas funciones de transferencia presentan la misma estructura de segundo orden, lo que permite considerar al circuito RLC como modelo eléctrico equivalente del sistema masa-resorte-amortiguador.")

# =================== Simulación Interactiva =====================
st.markdown("## 🧪 Simulación Interactiva")

# Sidebar: tipo de respuesta y parámetros
st.sidebar.header("⚙️ Parámetros del Sistema")
tipo_respuesta = st.sidebar.selectbox(
    "Tipo de Respuesta",
    ["Subamortiguada", "Sobreamortiguada", "Amortiguamiento Crítico", "Inestable"]
)

wn = st.sidebar.slider("Frecuencia Natural ωₙ (rad/s)", 0.1, 20.0, 5.0)
if tipo_respuesta == "Subamortiguada":
    zeta = st.sidebar.slider("ζ", 0.0, 1.0, 0.5)
elif tipo_respuesta == "Sobreamortiguada":
    zeta = st.sidebar.slider("ζ", 1.0, 2.0, 1.2)
elif tipo_respuesta == "Amortiguamiento Crítico":
    zeta = 1.0
else:
    zeta = st.sidebar.slider("ζ", -1.0, 0.0, -0.1)

# Parámetros físicos y eléctricos
m = 1.0
k = wn**2
c = 2 * zeta * wn * m
L = m
R = c
C = 1 / k

num = [1]
den = [m, c, k]
sys = TransferFunction(num, den)

# Checkbox para seleccionar lazo cerrado
lazo_cerrado = st.sidebar.checkbox("Simular en Lazo Cerrado", value=False)

# Redefinir sistema si es lazo cerrado
if lazo_cerrado:
    sys = feedback(sys, 1)

# Mostrar parámetros
col1, col2, col3 = st.columns(3)
col1.metric("Masa (m)", f"{m:.2f} kg")
col2.metric("Amortiguador (c)", f"{c:.2f} Ns/m")
col3.metric("Resorte (k)", f"{k:.2f} N/m")

col1, col2, col3 = st.columns(3)
col1.metric("Inductor (L)", f"{L:.2f} H")
col2.metric("Resistor (R)", f"{R:.2f} Ω")
col3.metric("Capacitor (C)", f"{C:.4f} F")

# Respuesta al escalón
t = np.linspace(0, 10, 1000)
t_step, y_step = step(sys, T=t)
st.markdown(f"### 🔄 Sistema en {'Lazo Cerrado' if lazo_cerrado else 'Lazo Abierto'}")

st.markdown("### 📈 Respuesta al Escalón")
fig1, ax1 = plt.subplots()
ax1.plot(t_step, y_step, label="Escalón")
ax1.set_xlabel("Tiempo (s)")
ax1.set_ylabel("Respuesta")
ax1.grid(True)
st.pyplot(fig1)

# Respuesta al impulso
t_impulse, y_impulse = impulse(sys, T=t)
st.markdown("### 💥 Respuesta al Impulso")
fig2, ax2 = plt.subplots()
ax2.plot(t_impulse, y_impulse, label="Impulso", color='orange')
ax2.set_xlabel("Tiempo (s)")
ax2.set_ylabel("Respuesta")
ax2.grid(True)
st.pyplot(fig2)

# Respuesta a la rampa
ramp_input = t
_, y_ramp, _ = lti(num, den).output(ramp_input, t)
st.markdown("### 🔼 Respuesta a la Rampa")
fig3, ax3 = plt.subplots()
ax3.plot(t, y_ramp, label="Rampa", color='green')
ax3.set_xlabel("Tiempo (s)")
ax3.set_ylabel("Respuesta")
ax3.grid(True)
st.pyplot(fig3)

# Mapa de polos y ceros
st.markdown("### ✳️ Mapa de Polos y Ceros")
poles = np.roots(den)
zeros = np.roots(num)
fig4, ax4 = plt.subplots()
ax4.scatter(np.real(poles), np.imag(poles), color='red', marker='x', label='Polos')
ax4.scatter(np.real(zeros), np.imag(zeros), color='blue', marker='o', label='Ceros')
ax4.axhline(0, color='black', lw=1)
ax4.axvline(0, color='black', lw=1)
ax4.set_xlabel("Re")
ax4.set_ylabel("Im")
ax4.grid(True)
ax4.legend()
st.pyplot(fig4)

# Parámetros temporales si subamortiguado
if 0 < zeta < 1:
    wd = wn * np.sqrt(1 - zeta**2)
    tr = np.pi / wd
    tp = np.pi / wd
    Mp = np.exp(-zeta * np.pi / np.sqrt(1 - zeta**2))
    ts = 4 / (zeta * wn)

    st.markdown("### 📊 Parámetros Temporales")
    df = pd.DataFrame({
        "Parámetro": [
            "Tiempo de levantamiento (tr)",
            "Tiempo al pico (tp)",
            "Sobreimpulso máximo (Mp)",
            "Tiempo de establecimiento (ts)"
        ],
        "Valor": [
            f"{tr:.3f} s",
            f"{tp:.3f} s",
            f"{Mp*100:.2f} %",
            f"{ts:.3f} s"
        ]
    })
    st.table(df)

st.markdown("### 📉 Diagrama de Bode")
w, mag, phase = bode(sys)
fig5, (ax5, ax6) = plt.subplots(2, 1, figsize=(8, 6))
ax5.semilogx(w, mag)
ax5.set_ylabel("Magnitud (dB)")
ax5.grid(True, which="both")

ax6.semilogx(w, phase)
ax6.set_xlabel("Frecuencia (rad/s)")
ax6.set_ylabel("Fase (°)")
ax6.grid(True, which="both")

st.pyplot(fig5)

```

```python
%%writefile 📻_Punto_2_Modelo.py

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

st.set_page_config(
    page_title="Punto 2 - Teoría SSB-AM",
    page_icon="📖",
    layout="wide"
)

st.title("📖 Punto 2: Teoría del Sistema SSB-AM")

st.markdown("""
## 📘 Introducción teórica

La modulación en **banda lateral única (SSB-AM, *Single Sideband Amplitude Modulation*)** es una técnica de comunicación eficiente que transmite **solo una de las dos bandas laterales** generadas al modular una señal por amplitud. Esto permite **reducir el ancho de banda a la mitad**, sin perder la información contenida en el mensaje original.

En la modulación AM convencional, la señal modulada se describe como:
""")

st.latex(r"s_{AM}(t) = [1 + m(t)] \cdot \cos(2\pi f_c t)")

st.markdown(r"""
Donde $ m(t) $ representa la **señal mensaje** y $ f_c $ es la **frecuencia de la portadora**.
Esta operación genera una señal con dos copias simétricas del espectro de $ m(t) $:
una centrada en $ +f_c $, conocida como **banda lateral superior (USB, *Upper Sideband*)**,
y otra centrada en $ -f_c $, llamada **banda lateral inferior (LSB, *Lower Sideband*)**.

En la modulación **SSB**, se transmite **solo una** de estas bandas, eliminando la otra mediante:

- La **transformada de Hilbert**, que permite construir la envolvente analítica, o
- Un **filtro pasa banda** que aísla la banda deseada.

---

## 🧮 Modelo matemático en el dominio del tiempo

Una de las formas más comunes de obtener una señal SSB es a través de la transformada de Hilbert.
La señal modulada resultante se expresa como:

""")

st.latex(r"s_{SSB}(t) = m(t) \cdot \cos(2\pi f_c t) + \hat{m}(t) \cdot \sin(2\pi f_c t)")

st.markdown(r"""
Aquí, $hat{m}(t)$ representa la **transformada de Hilbert** de $ m(t) $.
Dependiendo del signo, se obtiene:

- **USB (Upper Sideband)**: se usa el signo **positivo**.
- **LSB (Lower Sideband)**: se usa el signo **negativo**.

Este modelo evita transmitir redundancia espectral.

---

## 📊 Análisis en el dominio de la frecuencia

En el dominio de la frecuencia, es útil analizar primero una modulación tipo DSB-SC (*Double Sideband - Suppressed Carrier*), donde:

""")

st.latex(r"s_{DSB}(t) = m(t) \cdot \cos(2\pi f_c t)")

st.markdown("Aplicando la transformada de Fourier, se obtiene:")

st.latex(r"S_{DSB}(f) = \frac{1}{2} \left[ M(f - f_c) + M(f + f_c) \right]")

st.markdown(r"""
Esto indica que el espectro de la señal mensaje $ M(f) $ es **desplazado** a las frecuencias $ +f_c $ y $ -f_c $, generando dos bandas laterales simétricas.

Para obtener una señal SSB, se elimina **una de estas bandas laterales**. Por ejemplo:

- Si se conserva la banda en $ +f_c $, se obtiene la **USB**.
- Si se conserva la banda en $ -f_c $, se obtiene la **LSB**.

La expresión general del espectro en modulación SSB es:

""")

st.latex(r"S_{SSB}(f) = M(f - f_c) \quad \text{o bien} \quad M(f + f_c)")

st.markdown(r"""
Este desplazamiento se logra aplicando un **filtro pasa banda** adecuado o construyendo una **señal analítica compleja** con ayuda de la transformada de Hilbert.

---

## 🔄 Proceso completo

1. **Modulación**: se genera la señal SSB aplicando Hilbert o un filtro que elimine una de las bandas.
2. **Transmisión**: se envía únicamente la banda lateral deseada.
3. **Demodulación**: se multiplica la señal recibida por $$ \cos(2\pi f_c t) $$ y se filtra pasa bajos para recuperar el mensaje original:

""")

st.latex(r"r(t) = s_{SSB}(t) \cdot \cos(2\pi f_c t) \xrightarrow{\text{Filtro PB}} m(t)")

st.markdown("""
---

Esta formulación muestra cómo la modulación SSB-AM permite transmitir señales de manera más eficiente, al aprovechar principios fundamentales de la teoría de señales y de la transformada de Fourier.
""")
```

```python
%%writefile 📡_Punto_2_SSB_MAM.py


import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert, butter, filtfilt
from scipy.fft import fft, fftfreq, fftshift
import os
import tempfile
import subprocess
import yt_dlp as youtube_dl                                  # (yt-dlp ≫ youtube_dl)
import browser_cookie3
from scipy.io import wavfile

st.set_page_config(page_title="Punto 2 – SSB-AM",
                   page_icon="📡",
                   layout="wide")
st.title("📡 Punto 2 — Modulación y Demodulación SSB-AM")


st.markdown(
    """
En modulación **banda lateral única** (SSB-AM):

1. Partimos de la señal mensaje $m(t)$.
2. Calculamos su **Transformada de Hilbert** $\\hat m(t)$, que la gira $90^{\\circ}$.
3. Construimos la **señal analítica**
   $$
   m_a(t)=m(t)+j\\,\\hat m_a(t)
   $$
4. Desplazamos el espectro positivo a la portadora $f_c$ y tomamos la parte real:
   $$
   s_{\\text{SSB}}(t)=\\Re\\bigl\\{\,m_a(t)\\,e^{j2\\pi f_c t}\\bigr\\}
   $$
Así sólo queda **una banda lateral** y se ahorra la mitad del ancho de banda.
"""
)

st.header("Pulso rectangular")

# ---------- 1. Controles en la barra lateral ----------------
sidebar = st.sidebar
sidebar.subheader("⚙️ Parámetros del pulso")
dur = sidebar.slider("Duración (s)", 0.01, 1.00, 0.10, 0.01)
A   = sidebar.slider("Amplitud",      0.1,  5.0,  1.0,  0.1)
fc  = sidebar.slider("f\u2095 (Hz)",     10,   500,  100,  10)

# ---------- 2. Señal mensaje m(t) ---------------------------
Fs = 5_000                               # frecuencia de muestreo
t  = np.arange(-0.5, 0.5, 1/Fs)

m_t    = A * ((t >= -dur/2) & (t <= dur/2)).astype(float)

st.subheader("1️⃣ Mensaje $m(t)$ — Pulso rectangular")
st.markdown(fr"""
**Definición.** Generamos un pulso rectangular de amplitud $A={A}$
y duración $\tau={dur}\,\text{{s}}$ centrado en $t=0$:


""")

# ---------- 3. Transformada de Hilbert ----------------------
m_hilb = np.imag(hilbert(m_t))

st.subheader("2️⃣ Transformada de Hilbert $\\hat m(t)$")
st.markdown(r"""
Aplicamos la transformada de Hilbert para obtener una copia de $m(t)$
desfasada $90^{\circ}$.
Con $m(t)$ formará la señal analítica
$$
m_a(t)=m(t)+j\,\hat m(t),
$$
base de la modulación SSB.
""")

# ---------- 4. Señal SSB ------------------------------------
ssb = m_t*np.cos(2*np.pi*fc*t) - m_hilb*np.sin(2*np.pi*fc*t)

st.subheader("3️⃣ Construcción de $s_{\\text{SSB}}(t)$")
st.markdown(r"""
Combinamos $m(t)$ y $\hat m(t)$ con senos y cosenos a la portadora $f_c$:

$$
s_{USB}(t)=m(t)\cos 2\pi f_c t-\hat m(t)\sin 2\pi f_c t.
$$

Así conservamos **solo la banda lateral superior (USB)**.
""")

# ---------- 5. Espectro |S(f)| ------------------------------
freq = fftshift(fftfreq(len(ssb), d=1/Fs))
Sabs = fftshift(np.abs(fft(ssb)))

# ---------- 6. Demodulación ---------------------------------
y_det   = ssb * np.cos(2*np.pi*fc*t)      # detector productivo
fc_lp   = 2 / dur                         # corte ≈ ancho de m(t)
b_lp, a_lp = butter(5, fc_lp, btype='low', fs=Fs)
m_rec   = filtfilt(b_lp, a_lp, y_det)

st.subheader("4️⃣ Demodulación")
st.markdown(r"""
Multiplicamos por $\cos 2\pi f_c t$ (detector productivo) y filtramos con
un Butterworth pasa-bajas (orden 5, corte $\simeq 1/\tau$) para recuperar
el mensaje.
La curva sólida es $\hat m_{\text{rec}}(t)$; la punteada, el original.
""")

# ---------- 7. Gráficas (tiempo) ----------------------------
fig_time, axs = plt.subplots(4, 1, figsize=(10, 10), sharex=True)
axs[0].plot(t, m_t);              axs[0].set_title("m(t)")
axs[1].plot(t, m_hilb, 'orange'); axs[1].set_title("Ĥ{m}(t)")
axs[2].plot(t, ssb,  'green');    axs[2].set_title("s_SSB(t)")
axs[3].plot(t, m_rec, label="m̂(t)")
axs[3].plot(t, m_t, '--', alpha=0.6, label="m(t)")
axs[3].set_title("4️⃣ Demodulación"); axs[3].legend()
for ax in axs: ax.grid()
axs[-1].set_xlabel("t (s)")
st.pyplot(fig_time)

# ---------- 8. Gráfica (frecuencia) -------------------------
st.subheader("5️⃣ Espectro $|S(f)|$")
st.markdown("La FFT muestra energía solo en la banda lateral superior.")
fig_freq, axf = plt.subplots(figsize=(10, 3))
axf.plot(freq, Sabs)
axf.set_xlim(-2000, 2000)
axf.set_ylim(0, 1000)
axf.set_xlabel("Frecuencia (Hz)")
axf.set_ylabel("|S(f)|")
axf.set_title("Espectro de la señal SSB")
axf.grid()
st.pyplot(fig_freq)



# ---------- descarga y conversión a WAV ------------------------
YT_URL  = "https://www.youtube.com/watch?v=JtH68PJIQLE"
PREFIX  = "audio_default"

# ---- Descarga audio con yt-dlp -------------------------------
if not os.path.exists(f"{PREFIX}.mp3"):
    st.info("Descargando audio… ")
    try:
        cookies = browser_cookie3.chrome()
    except Exception:
        cookies = None
    ydl_opts = {
        "format": "bestaudio/best",
        "outtmpl": f"{PREFIX}.mp3",
        "quiet": True,
        "no_warnings": True,
    }
    if cookies:
        ydl_opts["cookiejar"] = cookies
    try:
        with youtube_dl.YoutubeDL(ydl_opts) as ydl:
            ydl.download([YT_URL])
        st.success("Descarga completada ✔️")
    except Exception as e:
        st.error(f"No se pudo descargar: {e}")

# ---- Conversión a WAV ----------------------------------------
if os.path.exists(f"{PREFIX}.mp3") and not os.path.exists(f"{PREFIX}.wav"):
    wav_tmp = tempfile.NamedTemporaryFile(suffix=".wav", delete=False).name
    try:
        subprocess.run(
            ["ffmpeg", "-y", "-i", f"{PREFIX}.mp3",
             "-ac", "1", "-ar", "48000", wav_tmp],
            capture_output=True, check=True
        )
        os.replace(wav_tmp, f"{PREFIX}.wav")
    except subprocess.CalledProcessError as err:
        st.error("ffmpeg error:\n" + err.stderr.decode())

# ---- Cargar y recortar audio ---------------------------------
if os.path.exists(f"{PREFIX}.wav"):
    Fs_a, audio = wavfile.read(f"{PREFIX}.wav")
    audio = audio.astype(float)
    if audio.ndim == 2:         # estéreo → mono
        audio = audio[:, 0]
    audio /= np.max(np.abs(audio))
    audio = audio[: int(Fs_a * 5)]   # 5 s
    t_a = np.linspace(0, len(audio)/Fs_a, len(audio))

    # ---- Reproducir sólo los 5 s recortados -----------------------
    trim_path = tempfile.NamedTemporaryFile(suffix=".wav", delete=False).name

# convierte de float [-1,1] a int16 antes de guardar
    wavfile.write(trim_path, Fs_a, (audio * 32767).astype(np.int16))

    st.audio(trim_path, format="audio/wav")

# ------------------------------------------------------------------
#  🔊 6. Procesamiento SSB-AM del segmento de audio (5 s)
# ------------------------------------------------------------------
st.header("🎵 Segmento de audio – Modulación y Demodulación SSB-AM")
st.markdown(r"""
## 🎧 Procesamiento completo de una señal de audio (5 segundos)

En esta sección aplicamos otra vez el proceso **completo de modulación y demodulación SSB-AM** (banda lateral única) sobre un fragmento de audio real (Canción: Oblivion - Grimes). Este flujo permite observar cómo una señal puede ser **transmitida eficientemente** ocupando solo la mitad del ancho de banda requerido por una modulación AM tradicional.

---

### 🔊 1. Preprocesamiento del audio

Antes de modular, es importante ajustar el audio:

- **Normalización:**
  $$ m_a(t) = \frac{\text{audio}}{\max(|\text{audio}|)} $$
  Esto asegura que la señal de mensaje quede entre −1 y 1, evitando saturación al modular.

- **Frecuencia de muestreo:**
  $$ F_s = 48\ \text{kHz} $$
  Esta tasa ofrece una buena resolución temporal y fidelidad para el audio.

- **Duración seleccionada:**
  Se trabaja con 5 segundos de señal, lo suficiente para ver la modulación sin sobrecargar el sistema.

---

### 🌀 2. Transformada de Hilbert

La **transformada de Hilbert** crea una versión desfasada \( \hat m_a(t) \) del mensaje:

$$ \hat m_a(t) = \mathcal{H}\{m_a(t)\} $$

- **Interpretación en tiempo:**
  Convolución con \( h(t) = \frac{1}{\pi t} \).

- **Interpretación en frecuencia:**
  Rota todas las frecuencias \( +90^\circ \) si son negativas, \( -90^\circ \) si son positivas.

- **¿Para qué sirve?**
  Permite generar una **señal analítica** que tiene solo frecuencias positivas (fundamental para SSB).

---

### 🔀 3. Generación de la señal analítica

La señal analítica se define como:

$$ m_{\mathrm{an}}(t) = m_a(t) + j\,\hat m_a(t) $$

Esta forma compleja de la señal **simplifica el proceso de modulación**, ya que concentra toda la energía espectral en la banda positiva.

---

### 📡 4. Modulación SSB (USB)

Para obtener la señal SSB (banda lateral superior), se usa:

$$
\begin{aligned}
s_{\mathrm{USB}}(t) &= \Re\left\{ m_{\mathrm{an}}(t)\cdot e^{j 2\pi f_c t} \right\} \\[6pt]
&= m_a(t)\cos(2\pi f_c t) - \hat m_a(t)\sin(2\pi f_c t)
\end{aligned}
$$

- **¿Por qué es útil?**
  Al eliminar la banda lateral inferior, se reduce el **ancho de banda a la mitad** comparado con AM tradicional.

---

### 🎯 5. Demodulación coherente (con portadora sincronizada)

Para recuperar la señal original:

1. **Multiplicamos la señal modulada con una portadora:**
   $$ y(t) = s_{\mathrm{USB}}(t)\cdot \cos(2\pi f_c t) $$

2. **Filtramos con un pasa bajas ideal:**
   Se elimina todo lo que esté fuera del rango de frecuencias del mensaje (e.g., corte en 5 kHz).

3. **Resultado:**
   $$ m_{\mathrm{rec}}(t) \approx \tfrac{1}{2} m_a(t) $$
   Se obtiene el mensaje original (con una pequeña atenuación).

---

### 📊 6. Gráficas diagnósticas del proceso

| Gráfico | ¿Qué se analiza? |
|--------|-------------------|
| \( m_a(t) \) y \( \hat m_a(t) \) | Relación de desfase entre el mensaje y su Hilbert |
| \( s_{\mathrm{USB}}(t) \) | Señal modulada: se ve la envolvente y la supresión de banda |
| \( m_{\mathrm{rec}}(t) \) vs \( m_a(t) \) | Qué tan bien se recuperó la señal |
| \( |S(f)| \) | Espectro: debe aparecer solo una banda (USB) |

---
""")

# -- Parámetros (en la misma sidebar para no marear al usuario)
sidebar.subheader("⚙️ Parámetros del audio")
fc_a = sidebar.slider("f\u2095 audio (Hz)", 300, 6000, 2000, 100)

# -- Señal mensaje m_a(t) = audio normalizado
st.subheader("Mensaje $m_{a}(t)$ — Audio (5 s)")

# Hilbert
audio_hilb = np.imag(hilbert(audio))

# SSB (banda lateral superior)
ssb_a = audio*np.cos(2*np.pi*fc_a*t_a) - audio_hilb*np.sin(2*np.pi*fc_a*t_a)

# Demodulación con detector productivo + mismo Butterworth
fc_lp_a = 0.9 * (Fs_a/2) if fc_a > Fs_a/2 else fc_a*0.4   # corte ≈ banda del mensaje
b_lp_a, a_lp_a = butter(5, fc_lp_a, btype='low', fs=Fs_a)
rec_a   = filtfilt(b_lp_a, a_lp_a, ssb_a*np.cos(2*np.pi*fc_a*t_a))

# ---------- Gráficas en el tiempo --------------------------------
fig_a_time, axs_a = plt.subplots(4,1,figsize=(10,10),sharex=True)
axs_a[0].plot(t_a, audio);            axs_a[0].set_title("mₐ(t)")
axs_a[1].plot(t_a, audio_hilb,'orange'); axs_a[1].set_title("Ĥ{mₐ}(t)")
axs_a[2].plot(t_a, ssb_a,  'green');  axs_a[2].set_title("s_SSB,audio(t)")
axs_a[3].plot(t_a, rec_a,   label="m̂ₐ(t)")
axs_a[3].plot(t_a, audio, '--', alpha=0.6, label="mₐ(t)")
axs_a[3].set_title("4️⃣ Demodulación audio"); axs_a[3].legend()
for ax in axs_a: ax.grid()
axs_a[-1].set_xlabel("t (s)")
st.pyplot(fig_a_time)

# ---------- Espectro ------------------------------------------------
st.subheader("5️⃣ Espectro del audio SSB $|S_a(f)|$")
freq_a = fftshift(fftfreq(len(ssb_a), d=1/Fs_a))
Sa_abs = fftshift(np.abs(fft(ssb_a)))

fig_a_freq, axfa = plt.subplots(figsize=(10,3))
axfa.plot(freq_a, Sa_abs)
axfa.set_xlim(-10000, 10000)
axfa.set_ylim(0, 5000)
axfa.set_xlabel("Frecuencia (Hz)")
axfa.set_ylabel("|S(f)|")
axfa.set_title("Espectro audio SSB")
axfa.grid()
st.pyplot(fig_a_freq)

# ------------------------------------------------------------------
# 7. Análisis del filtro IIR (Butterworth orden 5) – Bode y polos/ceros
# ------------------------------------------------------------------
import scipy.signal as sig

# ── Encabezado visible en la app ───────────────────────────────────
st.header("🧪 Filtro pasa-bajas – Bode y plano z")

st.markdown(r"""
## Filtro pasa-bajas

El filtro que usamos para demodular la señal es un **Butterworth pasa-bajas** de orden 5, diseñado para extraer el mensaje original luego de la modulación SSB-AM.

Para evaluar su desempeño analizamos:

---
## Diagrama de Bode

El **diagrama de Bode** representa cómo el filtro afecta diferentes frecuencias.
Consta de dos curvas:

- 📏 **Magnitud (en decibelios [dB])**
  Muestra cuánto se **amplifica o atenúa** cada componente de frecuencia:
  - 0 dB → la frecuencia pasa sin cambios.
  - Valores negativos → la señal se atenúa (se debilita).

- 📐 **Fase (en grados)**
  Indica el **retraso o adelanto temporal** (desfase) que el filtro introduce:
  - Desfases suaves ayudan a mantener la forma de onda original.

🧠 *Ambas curvas usan un eje de frecuencias logarítmico*, lo que permite ver con claridad tanto las bajas como las altas frecuencias.

---
## Plano $z$: Polos y Ceros

Este plano nos muestra la "firma digital" del filtro en el dominio discreto:

- ✕ **Polos**
  Son los puntos donde la ganancia tiende a infinito.
  - Si están **dentro del círculo unidad**, el filtro es **estable**.

- ◯ **Ceros**
  Son las frecuencias que el filtro **anula completamente**.

🔁 El plano z es esencial para analizar estabilidad y el tipo de atenuación que aplica el filtro.

---
## 🧪 Resumen de interpretación

| Herramienta     | ¿Qué nos dice?                                    |
|------------------|----------------------------------------------------|
| **Bode – Magnitud** | Qué frecuencias pasan y cuáles se atenúan         |
| **Bode – Fase**     | Cómo se desfasa la señal en función de la frecuencia |
| **Plano $z$**       | Estabilidad del filtro y comportamiento en frecuencia |

---
""")

# ── Respuesta en frecuencia (Bode) ────────────────────────────────
# Tomamos el filtro del pulso: coeficientes b_lp, a_lp
w, h = sig.freqz(b_lp, a_lp, worN=2048, fs=Fs)   # w en Hz
mag   = 20 * np.log10(np.abs(h))                 # magnitud → dB
phase = np.unwrap(np.angle(h)) * 180 / np.pi     # fase → grados

fig_bode, (axm, axp) = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

axm.plot(w, mag)
axm.set_title("Magnitud")
axm.set_ylabel("|H| [dB]")
axm.grid(True, which="both")

axp.plot(w, phase)
axp.set_title("Fase")
axp.set_ylabel("∠H [°]")
axp.set_xlabel("Frecuencia [Hz]")
axp.grid(True, which="both")

st.pyplot(fig_bode)

st.markdown(r"""
## 🔎 Interpretación del comportamiento del filtro

- ✅ La **magnitud** se mantiene cercana a 0 dB dentro de la banda pasante,
  lo que significa que las frecuencias del mensaje **se conservan sin atenuación**.

- 🔻 A partir de la frecuencia de corte, la magnitud **cae suavemente** a razón de aproximadamente \(-30\ \mathrm{dB/dec}\),
  lo que garantiza una **eliminación progresiva de componentes no deseados** (como las altas frecuencias de la portadora).

- 🔄 La **fase** varía de forma continua y sin saltos bruscos,
  lo que es típico de los filtros **Butterworth**, reconocidos por no distorsionar la forma de onda temporal del mensaje.

🧠 **Conclusión:** Este filtro cumple perfectamente con su objetivo:
recuperar el mensaje original evitando distorsión y manteniendo estabilidad.
""")

# ── Plano de polos y ceros ────────────────────────────────────────
z, p, _ = sig.tf2zpk(b_lp, a_lp)   # ceros (z) y polos (p) del filtro

fig_pz, axz = plt.subplots(figsize=(4, 4))

# círculo unidad (límite de estabilidad)
theta = np.linspace(0, 2*np.pi, 400)
axz.plot(np.cos(theta), np.sin(theta), 'k--', lw=0.5)

axz.scatter(np.real(z), np.imag(z), marker='o', label='Ceros')
axz.scatter(np.real(p), np.imag(p), marker='x', label='Polos')

axz.set_xlabel('Re{z}')
axz.set_ylabel('Im{z}')
axz.set_title("Plano z – Polos (✕) y Ceros (◯)")
axz.axis('equal')
axz.grid(True)
axz.legend()

st.pyplot(fig_pz)

st.markdown(r"""
Todos los polos aparecen **dentro** del círculo unidad ⇒
el filtro es **estable**.
Los ceros simétricos sobre el eje real provocan la atenuación de
altas frecuencias mostrada en el Bode.
""")

from scipy.io.wavfile import write

# Normalizar y convertir a int16
rec_a_int16 = np.int16(rec_a / np.max(np.abs(rec_a)) * 32767)

# Guardar como archivo temporal WAV
temp_audio = tempfile.NamedTemporaryFile(suffix=".wav", delete=False)
write(temp_audio.name, Fs_a, rec_a_int16)

# Mostrar reproductor en Streamlit
st.subheader("🔊 Reproducción del audio demodulado")
st.markdown("Escucha el resultado de la demodulación coherente SSB-AM:")
st.audio(temp_audio.name, format="audio/wav")
```
